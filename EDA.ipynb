```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.countplot(x=training_data['sum'], palette='coolwarm')
plt.xlabel('Number of Toxic Labels')
plt.ylabel('Count of Comments')
plt.title('Distribution of Toxic Comments')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
training_data['toxic'].value_counts().plot.pie(autopct='%1.1f%%', labels=['Non-Toxic', 'Toxic'], colors=['skyblue', 'red'])
plt.title('Proportion of Toxic vs. Non-Toxic Comments')
plt.ylabel('')
plt.show()
```

```python
training_data['comment_length'] = training_data['comment_text'].apply(len)

plt.figure(figsize=(8, 5))
sns.histplot(training_data[training_data['toxic'] == 0]['comment_length'], bins=50, color='blue', label='Non-Toxic', kde=True)
sns.histplot(training_data[training_data['toxic'] == 1]['comment_length'], bins=50, color='red', label='Toxic', kde=True)
plt.xlabel('Comment Length')
plt.ylabel('Count')
plt.title('Distribution of Comment Lengths')
plt.legend()
plt.show()
```
